```python
import pandas as pd
import google.generativeai as genai

# ---------------------
# YOUR CONFIG
# ---------------------
api = "AIzaSyBbOH8e90AbyCdr1bG_M1tiKoyjSf0Sjn8"
model_name = "gemini-2.5-flash"
qa_path = "/content/sample_data/chatbot_dataset.csv"

# Configure Gemini
genai.configure(api_key=api)
model = genai.GenerativeModel(model_name)

# ---------------------
# LOAD YOUR CSV DATA
# ---------------------
df = pd.read_csv(qa_path, encoding='latin1')

# Build context text from dataset
context_text = ""
for _, row in df.iterrows():
    context_text += f"Q: {row['question']}\nA: {row['answer']}\n"

# ---------------------
# GEMINI ANSWER FUNCTION
# ---------------------
def ask_gemini(question):
    prompt = f"""
You are Komalaa, wife of Dharani.
You are an admission and enquiry assistant for KGiSL.

Use ONLY the following information to answer:

{context_text}

User question: {question}
"""

    response = model.generate_content(prompt)
    return response.text


print("Welcome to the KGiSL Admission process!")
print("I am Komalaa, wife of Dharani. Ask me anything about admissions.")
print("Type 'exit' to quit.\n")

# ---------------------
# SIMPLE TERMINAL CHAT LOOP
# ---------------------
def chat():
    while True:
        user_input = input("You: ")
        if user_input.lower() == "exit":
            break
        try:
            response = ask_gemini(user_input)
            print("Komalaa:", response)
            print("-" * 50)
        except Exception as e:
            print("An error occurred:", e)

# ---------------------
# ATTACHED UI SECTION  (RUNS IN COLAB)
# ---------------------
!pip install gradio
import gradio as gr

def chatbot_ui(user_message):
    try:
        return ask_gemini(user_message)
    except Exception as e:
        return f"Error: {e}"

ui = gr.Interface(
    fn=chatbot_ui,
    inputs=gr.Textbox(lines=2, placeholder="Ask anything about admissions..."),
    outputs=gr.Textbox(lines=5),
    title="KGiSL Admission Chatbot - Komalaa",
    description="Ask Komalaa (wife of Dharani) your KGiSL admission questions ðŸ˜Š"
)

ui.launch()
```
